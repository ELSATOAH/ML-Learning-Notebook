## 1.Step getting data
First of all we need data, therefore we will take a dataset by the Data Professor's Github. for this we will use this dataset: https://github.com/dataprofessor/data/blob/master/delaney_solubility_with_descriptors.csv . 

press "Raw" and copy the URL, which should look like this: https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv .

To make a use of the dataset we need to read the CSV file.

df = dataframe


```python
import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

# 2. Step Splitting to X and y 
Here we are going to prepare our data by splitting it in 2 axis.
for this we define y for the last column, in this case 'logS'
and for X we remove the y axis 

```python
y=df['logS']
y
```

```python

X=df.drop('logS',axis=1)
X
```

# 3.Step Train and Test
We reached the Train and Test Section.
in general with working with datasets, we want to use 80% for training and 20% for testing.
and that is what we gonna do right here, we split the data in 2 parts and use them as training and testing material

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=100) #80% training and 20% testing data  #random_state=100 is used to fix the split
```

```python
X_train
```

```python
X_test
```

# 4.Step Model Building
In this part we build our model, with linear regression.
we define lr as our LinearRegression and fit it with our training data

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,y_train)
```

in this part we going to let the model make a prediction. the code should be straight forward
we define our prediction variables and let it predict by using training and test data

```python
y_lr_train_pred=lr.predict(X_train)
y_lr_test_pred=lr.predict(X_test)

print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)

lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
print('Linear Regression Train Mean Squared Error:',lr_train_mse)
print('Linear Regression Train R^2:',lr_train_r2)
print('Linear Regression Test Mean Squared Error:',lr_test_mse)
print('Linear Regression Test R^2:',lr_test_r2)
```

```python
lr_results=pd.DataFrame(['Linear Regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=['Model','Train MSE','Train R^2','Test MSE','Test R^2']
```

```python
lr_results
```

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(X_train,y_train)
```

```python
rf_train_pred=rf.predict(X_train)
rf_test_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse=mean_squared_error(y_train,rf_train_pred)
rf_train_r2=r2_score(y_train,rf_train_pred)
rf_test_mse=mean_squared_error(y_test,rf_test_pred)
rf_test_r2=r2_score(y_test,rf_test_pred)
```

```python
print('Random Forest Train Mean Squared Error:',rf_train_mse)
print('Random Forest Train R^2:',rf_train_r2)
print('Random Forest Test Mean Squared Error:',rf_test_mse)
print('Random Forest Test R^2:',rf_test_r2)
```

```python
rf_results=pd.DataFrame(['Random Forest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns=['Model','Train MSE','Train R^2','Test MSE','Test R^2']
rf_results
```

```python
df_model=pd.concat([lr_results,rf_results],axis=0)
df_model
```

```python
df_model.reset_index(drop=True,inplace=True)
df_model
```
